# Random Forest

```python
import pandas as pd

# Path to data file
parquet_file_path = '../../Costa_Rica_Data/Data Acquisition Output/extracted_gee_data/en_clean_gee_data.parquet'

# Load into a DataFrame
df = pd.read_parquet(parquet_file_path)

# Display head
df.head()
```

```python
# Check for columns with missing values
missing_columns = df.columns[df.isnull().any()]

# Display the results
print("Columns with missing values:", missing_columns)
```

```python
print(df.columns.tolist())
```

```python
y = df['Use']  # Set the target variable
X = df[['ndvi','evi','savi','brightness','wetness','slope','aspect','mTPI','ocs_1mMed', 'sand_1mMed','topDiv']]  # Predictor features

# Check the result
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
X.head()
```

```python
print(y.unique())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
y_prediction = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_prediction))
```

```python
# Save Model
import pickle

with open('rf_model.pkl', 'wb') as f:
    pickle.dump(rf, f)
```
